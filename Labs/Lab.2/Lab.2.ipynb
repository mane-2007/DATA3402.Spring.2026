*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
# Write your solution here
def game(board):
    n = len(board)

    for i in range(n):
        if board[i][0] != 0:
            same = True
            for j in range(n):
                if board[i][j] != board[i][0]:
                    same = False
            if same:
                return board[i][0]
            
    for j in range(n):
        if board[0][j] != 0:
            same = True
            for i in range(n):
                if board[i][j] != board[0][j]:
                    same = False
            if same:
                return board[0][j]
    
    if board[0][0] != 0:
        same = True
        for i in range(n):
            if board[i][i] != board[0][0]:
                same = False
        if same:
            return board[0][0]
    
    if board[0][n-1] != 0:
        same = True
        for i in range(n):
            if board[i][n-1-i] != board[0][n-1]:
                same = False
        if same:
            return board[0][n-1]
    
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                return -1
            
    return 0
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
# Write your solution here
def place_finder(board, player, row, col):
    if board[row][col] == 0:
        board[row][col] = player
        return True
    else:
        return False
````

*Exercise 7:* Write a function that takes a board, a player number, and a location string (as in Exercise 6), then uses your function from Exercise 5 to update the board.

````python
# Write your solution here
def moves(board, player, move):
    letters = ['A','B','C','D','E','F','G','H','I','J']
    row_character = move[0].upper()
    col_character = move[1]

    row = letters.index(row_character)
    col = int(col_character) - 1

    return place_finder(board, player, row, col)
````

````python
# Test your solution here
symbols = {0: ' ', 1: 'X', 2: 'O'}
board = my_board(3)

moves(board, 1, 'c1')
moves(board, 2, 'a3')
moves(board, 1, 'c2')
moves(board, 2, 'b3')
moves(board, 1, 'c3')
the_board(board, symbols)
````

*Exercise 8:* Write a function that is called with a board and player number, takes input from the player using Python's `input()`, and modifies the board using your function from Exercise 7.

Keep asking for input until the player enters a valid location that results in a valid move.


````python
# Write your solution here
def player_move(board, player):
    while True:
        move = input('Enter your move: ')
        move = move.strip()

        success = moves(board, player, move)

        if success:
            break
        else:
            print('Invalid move, try again')
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
# Write yourrr solution here
symbols = {0: ' ', 1: 'X', 2: 'O'}

def tictactoe_game(n=3):
    board = my_board(n)
    current_player = 1

    while True:
        labels_board(board, symbols)

        player_move(board, current_player)

        status = game(board)

        if status == -1:
            current_player = 2 if current_player == 1 else 1
        elif status == 0:
            labels_board(board, symbols)
            print("It's a draw!")
            break
        else:
            labels_board(board, symbols)
            print(f'Player {symbols[status]} wins!')
            break
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
# Write your solution here
def try_move(player, r, c, board=3):
    if board[r][c] != 0:
        return False
    board[r][c] = player
    win = game(board)
    board[r][c] = 0

    return win == player

def computer_moves(board, computer, human):
    n = len(board)
    for i in range(n):
        for j in range(n):
            if try_move(computer, i, j, board):
                return (i, j)
    
    for i in range(n):
        for j in range(n):
            if try_move(human, i, j, board):
                board[i][j] = computer
                return (i, j)
    
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = computer
                return (i, j)
            
def player_vs_computer(board, com_moves):
    player = 1
    computer = 2

    while True:
        labels_board(board, symbols)

        player_move(board, player)

        if game(board) == player:
            labels_board(board, symbols)
            print('You win')
            break
        
        if game(board) == 0:
            labels_board(board, symbols)
            print('Its a draw')
            break
        
        com_moves(board, computer, player)

        if game(board) == computer:
            labels_board(board, symbols)
            print('Computer wins')
            break
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
def score_board(board, depth, computer, human, turn):
    result = game(board)

    if result == computer:
        return 1
    if result == human:
        return -1
    if result == 0:
        return 0
    if depth == 0:
        return 0
    
    n = len(board)
    
    if turn == computer:
        best = -2
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = computer
                    val = score_board(board, depth-1, computer, human, human)
                    board[i][j] = 0
                    if val > best:
                        best = val
        return best
    else:
        best = 2
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = human
                    val = score_board(board, depth-1, computer, human, computer)
                    board[i][j] = 0
                    if val < best:
                        best = val
        return best
    
def computer_depth(board, computer, human, depth):
    n = len(board)
    best_score = -2
    move = None

    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = computer
                val = score_board(board, depth-1, computer, human, human)
                board[i][j] = 0
                if val > best_score:
                    best_score = val
                    move = (i, j)

    if move:
        board[move[0]][move[1]] = computer

def computer_depth_move(board, computer, human):
    computer_depth(board, computer, human, 9)
def computer_depth_move3(board, computer, human):
    computer_depth(board, computer, human, 4)
````
