1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, max_values):
        self.count = 0
        self.max_values = max_values

    def increment(self):
        if self.count < self.max_values:
            self.count += 1
        else:
            print('Maximum count reached')
    
    def reset(self):
        self.count = 0
        print('Reset to 0')
    
    def get_value(self):
        return self.count
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, length):
        self.canvas = paint.Canvas(width, length)
        self.shapes = []

    def add_shape(self, shape, char='o'):
        self.shapes.append((shape, char))

    def render(self):
        self.canvas.clear_canvas()
        for shape, char in self.shapes:
            shape.paint(self.canvas, char=char)
        self.canvas.display()

    def move_shape(self, index, a, b):
        if 0 <= index < len(self.shapes):
            shape, char = self.shapes[index]
            print(f'Moving shape from {index} to ({a}, {b})')
```

12. Implement the ability to load/save raster drawings and demonstate that your method works. One way to implement this ability:

   * Overload `__repr__` functions of all objects to return strings of the python code that would construct the object.
   
   * In the save method of raster drawing class, store the representations into the file.
   * Write a loader function that reads the file and uses `eval` to instantiate the object.

For example:

```python
#The __repr__ was done inside the paint.py file
class RasterDrawing:
    def __init__(self, width, length):
        self.canvas = paint.Canvas(width, length)
        self.shapes = []
        self.width = width
        self.length = length

    def add_shape(self, shape, char='o'):
        self.shapes.append((shape, char))

    def render(self):
        self.canvas.clear_canvas()
        for shape, char in self.shapes:
            shape.paint(self.canvas, char=char)
        self.canvas.display()

    def move_shape(self, index, a, b):
        if 0 <= index < len(self.shapes):
            shape, char = self.shapes[index]
            print(f'Moving shape from {index} to ({a}, {b})')

    def __repr__(self):
        return f'RasterDrawing({repr(self.width)}, {repr(self.length)})'

    def save(self, filename):
        f = open(filename, 'w')
        data_to_save = [self.width, self.length, self.shapes]
        f.write(repr(data_to_save))
        f.close()
        
    @staticmethod
    def load(filename):
        f = open(filename, 'r')
        tmp = eval(f.read())
        f.close()

        new_drawing = RasterDrawing(tmp[0], tmp[1])
        new_drawing.shapes = tmp[2]
        return new_drawing
```
